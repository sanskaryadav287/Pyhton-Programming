1. Integer Variable

```python
age = 25
print(age)

#Stores a whole number.
```

2. Float Variable

```python
temperature = 36.6
print("Temperature:", temperature)

#Stores a decimal value.
```

3. String Variable

```python
name = "Gaurav"
print("Name:", name)

#Stores text inside quotes.
```

4. Boolean Variable

```python
is_student = True
print("Is Student:", is_student)

#Stores True or False.
```

5. List Variable

```python
fruits = ["apple", "banana", "cherry"]
print("Fruits:", fruits)

#Stores multiple items in one variable.
```

6. Tuple Variable

```python
coordinates = (10, 20)
print("Coordinates:", coordinates)

#Similar to list but immutable (cannot be changed).
```

7. Dictionary Variable

```python
student = {"name": "Arun", "age": 22, "grade": "A"}
print("Student:", student)

#Stores data in key-value pairs.
```

8. Set Variable

```python
unique_numbers = {1, 2, 3, 4, 4}
print("Unique Numbers:", unique_numbers)

#Stores unique values only.
```

9. NoneType Variable

```python
value = None
print("Value:", value)

#Represents no value (null).
```

10. Variable Reassignment

```python
x = 10
print("x:", x)
x = "Now I am a string"
print("x:", x)

#Python variables can change type dynamically.
```

```python
Text Type:	str
Numeric Types:	int, float, complex
Sequence Types:	list, tuple, range
Mapping Type:	dict
Set Types:	set, frozenset
Boolean Type:	bool
Binary Types:	bytes, bytearray, memoryview
None Type:	NoneType
```
