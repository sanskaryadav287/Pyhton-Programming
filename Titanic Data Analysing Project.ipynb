```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Titanic-Dataset.csv")

# Cleaning
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.drop('Cabin', axis=1, inplace=True)

# Convert categories
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# ----------------------------------------------------------
# 1️⃣ Survival Count
plt.figure()
df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Count")
plt.xlabel("Survived (0=No, 1=Yes)")
plt.ylabel("Count")
plt.show()

# ----------------------------------------------------------
# 2️⃣ Gender Distribution
plt.figure()
df['Sex'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Sex (0=Male, 1=Female)")
plt.ylabel("Count")
plt.show()

# ----------------------------------------------------------
# 3️⃣ Passenger Class Distribution
plt.figure()
df['Pclass'].value_counts().plot(kind='bar')
plt.title("Passenger Class Distribution")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()

# ----------------------------------------------------------
# 4️⃣ Survival Rate by Gender
plt.figure()
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.xlabel("Sex (0=Male, 1=Female)")
plt.ylabel("Survival Rate")
plt.show()

# ----------------------------------------------------------
# 5️⃣ Survival Rate by Pclass
plt.figure()
df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Pclass")
plt.ylabel("Survival Rate")
plt.show()

# ----------------------------------------------------------
# 6️⃣ Age Distribution Histogram
plt.figure()
df['Age'].plot(kind='hist', bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# ----------------------------------------------------------
# 7️⃣ Fare Distribution Histogram
plt.figure()
df['Fare'].plot(kind='hist', bins=20)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()

# ----------------------------------------------------------
# 8️⃣ Age vs Survival (Boxplot)
plt.figure()
df.boxplot(column='Age', by='Survived')
plt.title("Age Distribution by Survival")
plt.suptitle("")  # removes extra title
plt.xlabel("Survived (0=No, 1=Yes)")
plt.ylabel("Age")
plt.show()

# ----------------------------------------------------------
# 9️⃣ Pclass vs Fare (Boxplot)
plt.figure()
df.boxplot(column='Fare', by='Pclass')
plt.title("Fare Distribution by Pclass")
plt.suptitle("")
plt.xlabel("Pclass")
plt.ylabel("Fare")
plt.show()

# ----------------------------------------------------------
# 🔟 Embarked vs Survival
plt.figure()
df.groupby('Embarked')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Embarkation Point")
plt.xlabel("Embarked (0=S, 1=C, 2=Q)")
plt.ylabel("Survival Rate")
plt.show()
```
