###  Total bill format ###

```python
purchase_amount=float(input("Enter Purchase Amount "))

if purchase_amount>=5000:
    dis_per=20
elif purchase_amount>=3000 and purchase_amount<5000:
    dis_per=10
else:
    print("No Discount")

dis_amount=(purchase_amount*dis_per)/100
amount_after_dis=purchase_amount-dis_amount

gst_percent=18
gst_amount=(amount_after_dis*gst_percent)/100

final_amount=amount_after_dis+gst_amount

print("\n==============Bill Summary=========")
print(f"Purchase amount       :u20B9{purchase_amount:.2f}")
print(f"Discount Applied      :{dis_per}%")
print(f"Discount Amount       :u20B9{dis_amount:.2f}")
print(f"Amount After Discount :u20B9{amount_after_dis:.2f}")
print(f"GST  ({gst_percent}%) :u20B9{gst_amount:.2f}")
print("-------------------------------------------------------")
print(f"Final Payable amount  :u20B9{final_amount:.2f}")
print("=======================================================")
print("Thank You for shoppng with us ! \u0001F602" )
```
