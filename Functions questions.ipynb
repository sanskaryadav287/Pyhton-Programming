# i) Write a Python program to check if the given number is prime or not? 


```python
num = int(input("Number : "))
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(f"{num} prime number")
else:
    print(f"{num} not a prime number")
```

# (ii) Write a program to check for  Palindrome, Armstrong & Fibonacci Series

```python

def check_palindrome(n):
    return str(n) == str(n)[::-1]


def check_armstrong(n):
    order = len(str(n))
    sum = 0
    temp = n
    while temp > 0:
        digit = temp % 10
        sum += digit ** order
        temp //= 10
    return sum == n


def check_fibonacci(n):
    a, b = 0, 1
    while b < n:
        a, b = b, a + b
    return b == n


number = int(input("n: "))

if check_palindrome(number):
    print(f"{number}Palindrome ")
else:
    print(f"{number} not a Palindrome")

if check_armstrong(number):
    print(f"{number} Armstrong ")
else:
    print(f"{number} not a Armstrong")

if check_fibonacci(number):
    print(f"{number} yes part of Fibonacci")
else:
    print(f"{number} not a part of Fibonacci")
```

# (iii) Create a Python function that contains a function to calculate the factorial of a number usingÂ recursion.

```python

def factorial(n):
    return 1 if n <= 1 else n * factorial(n - 1)

```
